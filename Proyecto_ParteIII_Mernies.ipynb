```python


# Cargar el dataset desde la URL especificada, ignorando la primera columna si es un índice
df = pd.read_csv('https://raw.githubusercontent.com/merniesgugh/CoderHouse-DS/main/kl.csv', encoding='latin1', index_col=0)

# Mantener solo las columnas numéricas
numerical_data = df.select_dtypes(include='number')

# Manejar valores faltantes eliminando filas con NaN en las columnas numéricas
numerical_data = numerical_data.dropna()

# Separar las variables predictoras y la variable objetivo
y = numerical_data['Overall']  # Variable objetivo
X = numerical_data.drop(columns=['Overall'])  # Variables predictoras

# Añadir una columna constante para la intercepción
X = sm.add_constant(X)

# Función para backward selection
def backward_selection(X, y, significance_level=0.05):
    initial_vars = X.columns.tolist()
    while len(initial_vars) > 0:
        model = sm.OLS(y, X[initial_vars]).fit()
        # Obtener la p-valor máximo entre las variables actuales
        max_p_value = max(model.pvalues)
        if max_p_value > significance_level:
            excluded_var = model.pvalues.idxmax()
            initial_vars.remove(excluded_var)
        else:
            break
    return initial_vars

# Ejecutar la selección hacia atrás
selected_vars = backward_selection(X, y)

# Remover 'const' si está en las variables seleccionadas
if 'const' in selected_vars:
    selected_vars.remove('const')

# Calcular la correlación con 'Overall' para las variables seleccionadas y ordenarlas
correlation_with_overall = numerical_data[selected_vars].corrwith(numerical_data['Overall']).abs()
correlation_sorted = correlation_with_overall.sort_values(ascending=False)

# Mostrar la lista ordenada de variables seleccionadas
print("Variables seleccionadas y su correlación con Overall, ordenadas de mayor a menor:")
for var, corr in correlation_sorted.items():
    print(f"{var}: {corr:.4f}")

# Graficar la correlación de las variables seleccionadas con 'Overall'
correlation_sorted.plot(kind='bar', title='Correlación con Overall')
plt.xlabel('Variables')
plt.ylabel('Correlación')
plt.show()


```

### Algunos Conceptos y Definiciones Previas:

### Uso de MSE, MAE y R² en el Análisis de Modelos de Regresión

Estos tres conceptos – el **Error Cuadrático Medio (MSE)**, el **Error Absoluto Medio (MAE)** y el **Coeficiente de Determinación (R²)** – son métricas clave para evaluar la precisión y eficacia de modelos de regresión:


### Error Cuadrático Medio (MSE)
El **Error Cuadrático Medio (MSE)** mide el promedio de los cuadrados de los errores, es decir, la diferencia entre los valores predichos por el modelo y los valores reales observados. Es una métrica de evaluación común en modelos de regresión que penaliza más los errores grandes. Un MSE más bajo indica un modelo con mejor precisión. Su fórmula es:

$$
\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

donde *yi* es el valor real, *ŷi* es el valor predicho, y *n* es el número de observaciones.

---

### Error Absoluto Medio (MAE)
El **Error Absoluto Medio (MAE)** es la media de los valores absolutos de las diferencias entre los valores predichos y los valores reales. A diferencia del MSE, el MAE no penaliza tanto los errores grandes, ya que no eleva al cuadrado las diferencias. Es una métrica útil cuando se quiere entender la magnitud promedio de los errores sin dar más peso a los errores mayores. Su fórmula es:

$$
\text{MAE} = \frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y}_i|
$$

donde *yi* es el valor real, *ŷi* es el valor predicho, y *n* es el número de observaciones.

---

### Coeficiente de Determinación (R²)
El **Coeficiente de Determinación (R²)**, también conocido como R cuadrado, mide la proporción de la varianza en la variable dependiente que es explicada por el modelo. Un valor de R² cercano a 1 indica que el modelo explica bien la variabilidad de los datos, mientras que un valor cercano a 0 sugiere que el modelo no es un buen predictor. Su fórmula es:

$$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$$

donde *yi* es el valor real, *ŷi* es el valor predicho, *ȳ* es la media de los valores reales, y *n* es el número de observaciones. Un R² alto sugiere que el modelo captura bien la tendencia de los datos.

A continuación, se presenta una comparación del rendimiento del modelo predictivo de Random Forest utilizando dos enfoques de validación: una división convencional en Train/Test y, por otro lado, una validación cruzada mediante K-Fold. Este último método no solo permite una evaluación más robusta del modelo, sino que también ayuda a reducir el riesgo de sobreajuste, proporcionando una estimación más confiable del rendimiento en datos no vistos.

```python

# Filtrar las columnas de interés (Overall y las seleccionadas)
data_filtered = df[['Overall', 'Reactions', 'Composure', 'Potential', 'Special', 'ShortPassing']].dropna()

# Definir las variables X (predictoras) y y (objetivo)
X = data_filtered[['Reactions', 'Composure', 'Potential', 'Special', 'ShortPassing']]  # Variables predictoras seleccionadas
y = data_filtered['Overall']  # Calificación general como variable objetivo

# Normalizar las variables predictoras
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crear el modelo de Random Forest
model = RandomForestRegressor(n_estimators=100, random_state=42)

# ---- Evaluación sin K-Fold (Train-Test Split) ----
# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Entrenar y evaluar el modelo
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse_single = mean_squared_error(y_test, y_pred)
mae_single = mean_absolute_error(y_test, y_pred)
r2_single = r2_score(y_test, y_pred)

# ---- Evaluación con K-Fold Cross-Validation ----
# Configuración de K-Fold Cross-Validation de 5 particiones
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Definir las métricas para la validación cruzada
mse_scorer = make_scorer(mean_squared_error)
mae_scorer = make_scorer(mean_absolute_error)
r2_scorer = make_scorer(r2_score)

# Ejecutar validación cruzada y calcular el promedio de las métricas
mse_scores = cross_val_score(model, X_scaled, y, cv=kf, scoring=mse_scorer)
mae_scores = cross_val_score(model, X_scaled, y, cv=kf, scoring=mae_scorer)
r2_scores = cross_val_score(model, X_scaled, y, cv=kf, scoring=r2_scorer)

mse_kfold = np.mean(mse_scores)
mae_kfold = np.mean(mae_scores)
r2_kfold = np.mean(r2_scores)

# ---- Comparación de Resultados ----
# Imprimir resultados
print("Resultados sin K-Fold (Train-Test Split):")
print(f" - Error cuadrático medio (MSE): {mse_single}")
print(f" - Error absoluto medio (MAE): {mae_single}")
print(f" - Coeficiente de determinación (R²): {r2_single}")

print("\nResultados con K-Fold Cross-Validation:")
print(f" - Error cuadrático medio (MSE): {mse_kfold}")
print(f" - Error absoluto medio (MAE): {mae_kfold}")
print(f" - Coeficiente de determinación (R²): {r2_kfold}")

# ---- Visualización ----
# Crear gráficos de barras para comparar los resultados
metrics = ['MSE', 'MAE', 'R²']
single_scores = [mse_single, mae_single, r2_single]
kfold_scores = [mse_kfold, mae_kfold, r2_kfold]

x = np.arange(len(metrics))  # Posiciones para las métricas
width = 0.35  # Ancho de las barras

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, single_scores, width, label='Train-Test Split')
bars2 = ax.bar(x + width/2, kfold_scores, width, label='K-Fold Cross-Validation')

# Añadir etiquetas y título
ax.set_xlabel('Métricas')
ax.set_ylabel('Valores')
ax.set_title('Comparación de Métricas: Train-Test Split vs. K-Fold Cross-Validation')
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Mostrar los valores encima de las barras
for bar in bars1 + bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.show()
```


**1. Error Cuadrático Medio (MSE):**
Train-Test Split: El MSE es de aproximadamente 6.38.
K-Fold Cross-Validation: El MSE promedio es de 6.23.

El MSE obtenido con K-Fold Cross-Validation es ligeramente menor, lo que sugiere que la validación cruzada proporciona un modelo un poco más preciso en términos de error cuadrático promedio. Este ligero descenso en el error indica una reducción en la variabilidad del modelo y un mejor rendimiento en general.

**2. Error Absoluto Medio (MAE):**
Train-Test Split: El MAE es de 1.82.
K-Fold Cross-Validation: El MAE promedio es de 1.80.

El MAE es muy similar entre ambos métodos, con una leve mejora al utilizar K-Fold Cross-Validation. Esto significa que, en promedio, los errores de predicción en ambas evaluaciones son casi iguales. Sin embargo, la estabilidad proporcionada por la validación cruzada permite tener una mejor estimación de este error en diferentes subconjuntos.

**3. Coeficiente de Determinación (R²):**
Train-Test Split: El R² es de 0.87.
K-Fold Cross-Validation: El R² promedio es también de 0.87.

El R² es prácticamente idéntico en ambos métodos, lo que indica que el modelo mantiene su capacidad de explicar aproximadamente el 87% de la variabilidad de Overall con las variables predictoras seleccionadas en ambos enfoques.

**Conclusión de la Comparación:**
La validación cruzada K-Fold proporciona resultados muy similares a los obtenidos con el método Train-Test Split, pero con la ventaja de una mayor estabilidad y fiabilidad. Al usar K-Fold Cross-Validation, se reduce la variabilidad de las métricas de rendimiento debido a que se evalúa el modelo en múltiples particiones, lo que hace que el promedio de los resultados sea una mejor representación de su rendimiento general. Aunque las diferencias en los valores de MSE, MAE y R² son pequeñas, el modelo validado con K-Fold es más confiable, ya que se ha probado en distintas configuraciones de datos.

**Reducción del Riesgo de Sobreajuste (Overfitting)**
Cuando un modelo se ajusta a una única partición de entrenamiento, puede aprender características específicas de esa partición, lo que lleva a sobreajuste. Con K-Fold Cross-Validation, el modelo se entrena y evalúa en varias configuraciones de los datos, reduciendo la probabilidad de que aprenda peculiaridades de un solo conjunto de entrenamiento.
Esto también ayuda a detectar si el modelo tiene una alta variabilidad en sus predicciones, lo cual es indicativo de un posible sobreajuste.
